**1.** Create a DataFrame `df` from this dictionary `data` which has the index `labels`.

````python
import pandas as pd
import numpy as np
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
print(df)
````

````python
#Diabetes model
df = pd.read_csv('/content/diabetes_model.csv')
df
````

````python
#Histogram for Univariate Analysis
plt.hist(df['Glucose'], bins=10, color="violet",edgecolor="black")
plt.xlabel('Glucose')
plt.ylabel('Level')
plt.title('Histogram of Glucose')
plt.show()
````

The histogram shows the distribution of glucose levels.It shows a normal distribution. Skewed bars indicate a non-normal distribution.The peak of the histogram represents the most common glucose level.

````python
plt.hist(df['BloodPressure'], bins=10, color="orange",edgecolor="Black")
plt.xlabel('Blood Pressure')
plt.ylabel('Range')
plt.title('Histogram of Blood Pressure')
plt.show()
````

````python
plt.scatter(df['Age'], df['BMI'], color="red")
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Scatter Plot of Glucose vs Blood Pressure')
plt.show()
````

The scatter plot shows the relationship between age and BMI (Body Mass Index).As age increases, there seems to be a slight upward trend in BMI. This shows that older individuals tend to have slightly higher BMIs.

The relationship is not very strong, as the points are scattered rather
than forming a clear linear pattern.
